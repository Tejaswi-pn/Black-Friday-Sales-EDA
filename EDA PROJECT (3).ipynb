```python
Q1 = df['Purchase'].quantile(0.25)
Q3 = df['Purchase'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

<h1>Univariate Analysis Observations</h1>

Univariate analysis focuses on examining one variable at a time.

The Gender distribution shows that male customers are higher in number than female customers.

Most customers belong to the age group 26–35 years.

City Category B has the highest number of customers, followed by Category C and A.

The Purchase Amount distribution is positively skewed, indicating that most purchases are of medium value with a few high-value transactions.

A large portion of customers are married compared to unmarried customers.

```python
cat_counts = (
    df[["Product_Category_1","Product_Category_2","Product_Category_3"]]
    .melt(value_name="category")["category"]
    .value_counts()
    .head(15)
)

plt.figure(figsize=(12,5))
cat_counts.plot(kind="bar")
plt.title("Most Frequent Product Categories (top 15)")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df["Product_ID"].value_counts().head(15)

sns.barplot(x=top_products.index, y=top_products.values)
plt.xticks(rotation=45)
plt.title("Top 15 Sold Products")
plt.show()
```

<h1>Bivariate Analysis Observations</h1>

Bivariate analysis examines the relationship between two variables.

Male customers spend more on average than female customers.

Customers aged 26–35 contribute the highest purchase amount compared to other age groups.

Married customers have slightly higher spending than unmarried customers.

Customers from City Category B and C show higher purchase values than Category A.

Purchase amount varies significantly across different occupations, indicating income influence.

Age and marital status together affect spending patterns.

<h1>MultiVariate Analysis</h1>

```python
cols = ["Purchase", "Product_Category_1", "Product_Category_2", "Product_Category_3", "Occupation"]

plt.figure(figsize=(10,6))
sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Selected Features)")
plt.show()
```

```python
cat_corr = df[["Product_Category_1","Product_Category_2","Product_Category_3","Purchase"]].corr()

sns.heatmap(cat_corr, annot=True, cmap="viridis")
plt.title("Product Categories Correlation")
plt.show()
```

```python
age_gender = df.groupby(["Age","Gender"])["Purchase"].mean().reset_index()

plt.figure(figsize=(12,6))
sizes = age_gender["Purchase"] / 20

plt.scatter(
    age_gender["Age"],
    age_gender["Purchase"],
    s=sizes,
    alpha=0.6,
    c=pd.factorize(age_gender["Gender"])[0]
)

plt.title("Average Purchase by Age & Gender (Bubble Size = Spend)")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df, x="Age", y="Purchase", hue="Gender")
plt.title("Spend Variability by Age & Gender")
plt.show()
```

<h1>Multivariate Analysis Observations</h1>

Multivariate analysis studies relationships among multiple variables simultaneously.

Male customers in the 26–35 age group from City Category B contribute the highest sales.

Purchase amount is influenced by a combination of age, occupation, and city category.

Married customers in mid-age groups tend to spend more than others.

High purchase amounts are commonly observed among customers with stable occupations and urban city categories.

Customer demographics collectively play a major role in determining purchasing behavior.
